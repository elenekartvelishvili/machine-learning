```python
from numpy import float_power
from sklearn.datasets import make_classification

x,y=make_classification(n_samples=5000,n_features=30,n_informative=25,n_redundant=5,weights=[0.8],flip_y=0.0,random_state=1)

from collections import Counter

Counter(y)
```

```python
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler

from imblearn.pipeline import Pipeline
sampler=Pipeline(steps=[('over',SMOTE(sampling_strategy=0.7)),('under',RandomUnderSampler(sampling_strategy=0.8))])

x,y=sampler.fit_resample(x,y)


Counter(y)
```
